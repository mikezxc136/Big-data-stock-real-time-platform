```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create figure and axes
fig, ax = plt.subplots(figsize=(12, 8))

# Disable axis
ax.axis('off')

# Create boxes
boxes = {
    "Technical Users": (0.1, 0.6, 0.3, 0.2),
    "Business Users": (0.6, 0.6, 0.3, 0.2),
    "Analyst Users": (0.1, 0.1, 0.3, 0.2),
    "Monitoring Users": (0.6, 0.1, 0.3, 0.2),
    "Central System": (0.35, 0.4, 0.3, 0.2)
}

# Add boxes to plot
for key, (x, y, w, h) in boxes.items():
    ax.add_patch(patches.FancyBboxPatch((x, y), w, h, boxstyle="round,pad=0.1", edgecolor='black', facecolor='lightgray'))
    ax.text(x + w/2, y + h/2, key, ha='center', va='center', fontsize=12, weight='bold')

# Add arrows
arrows = [
    ((0.25, 0.6), (0.5, 0.4)),
    ((0.75, 0.6), (0.65, 0.4)),
    ((0.25, 0.3), (0.5, 0.4)),
    ((0.75, 0.3), (0.65, 0.4))
]

for (x1, y1), (x2, y2) in arrows:
    ax.annotate("", xy=(x2, y2), xytext=(x1, y1), arrowprops=dict(arrowstyle="->", lw=2))

# Descriptions for each user group
descriptions = {
    "Technical Users": [
        "Manage and maintain the system",
        "Monitor for errors and troubleshoot issues",
        "Configure and optimize system performance",
        "Ensure LSTM model is updated and retrained"
    ],
    "Business Users": [
        "Access and analyze financial data",
        "Make informed business decisions",
        "Utilize LSTM model forecasts for planning"
    ],
    "Analyst Users": [
        "Perform in-depth data analysis",
        "Conduct complex queries",
        "Develop predictive models",
        "Evaluate and enhance LSTM model"
    ],
    "Monitoring Users": [
        "Monitor system health using Grafana",
        "Respond to alerts from Prometheus",
        "Monitor LSTM model predictions accuracy"
    ]
}

# Add descriptions to plot
for key, lines in descriptions.items():
    x, y, w, h = boxes[key]
    text = "\n".join(lines)
    ax.text(x + w/2, y - 0.05, text, ha='center', va='top', fontsize=10, wrap=True)

plt.show()
```
